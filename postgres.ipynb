```python
import psycopg2
```

```python
from dotenv import load_dotenv, find_dotenv
import os
dotenv_path = find_dotenv()
load_dotenv(dotenv_path)
password = os.getenv("PASSWORD")
```

```python
connection = psycopg2.connect(host="localhost", dbname='wonder_books', user = "postgres", port = 5432, password = password)
cursor = connection.cursor()
cursor.execute("""CREATE TABLE IF NOT EXISTS books(
               id BIGSERIAL NOT NULL PRIMARY KEY,
               name VARCHAR(255),
               author VARCHAR(255),
               format VARCHAR(255),
               price VARCHAR(255),
               link VARCHAR(255)
               );
""")
connection.commit()
cursor.close()
connection.close()
```
